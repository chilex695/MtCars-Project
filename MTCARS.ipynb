```python
Q1=df['hp'].quantile(0.25)
Q3=df['hp'].quantile(0.75)
IQR= Q3 - Q1

Lower_Bound = Q1-1.5*(IQR)
Upper_Bound =Q3+1.5*(IQR)

df = df[(df['hp']>=Lower_Bound) & (df['hp']<=Upper_Bound)]
```
